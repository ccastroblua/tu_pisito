#### Let's get only neighborhood and districts:

#### Districts:

```python
cols = ["neighborhood", "district"]
```

```python
distritos = list(barrios.district.unique())
```

```python
distritos
```

#### Neighborhoods:

```python
lista = []
```

```python
listado_barrios = barrios[cols].neighborhood.unique().tolist()
```

```python
listado_barrios
```

```python
geolocator = Nominatim(user_agent="tu_pisito", timeout = 5)
lista = []

for barrio in listado_barrios:
    neighborhood = barrio + ", Madrid"
    
    location = geolocator.geocode(neighborhood)
    try:    
        lat = location.latitude
        lon = location.longitude
    except:
        lat = 0
        lon = 0
        print(f"{barrio} no tiene lat y lon")
    
    dic = {}
    
    dic["neighborhood"] = barrio
    dic["latitude"] = lat
    dic["longitude"] = lon
    
    lista.append(dic)
```

#### Neighborhoods list:

```python
len(lista)
```

```python
new_df = pd.DataFrame(lista)
```
